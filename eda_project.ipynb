```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("student_performance_data.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['GPA'], bins=20, kde=True)

plt.title("Distribution of GPA")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='StudyTimeWeekly',
    y='GPA',
    data=df
)

plt.title("Study Time vs GPA")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='Absences',
    y='GPA',
    data=df
)

plt.title("Absences vs GPA")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='ParentalSupport',
    y='GPA',
    data=df
)

plt.title("Parental Support vs GPA")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='GradeClass',
    data=df
)

plt.title("Distribution of Grade Classes")

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
activity_gpa = df.groupby('Sports')['GPA'].mean()

print(activity_gpa)
```

```python
activity_gpa.plot(kind='bar')

plt.title("Average GPA by Sports Participation")

plt.show()
```
